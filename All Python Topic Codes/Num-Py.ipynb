# Numpy Array Basic Indexing & slicing

```python
arr2d=np.arange(49).reshape(7,7)
```

```python
arr2d
```

```python
arr2d[4,5]
```

```python
arr2d[5][6]
```

```python
arr3d=np.arange(120).reshape(3,20,2)
```

```python
arr3d
```

```python
arr1d=np.arange(15)
arr1d
```

```python
sliced=arr1d[4:10:2]
sliced
```
